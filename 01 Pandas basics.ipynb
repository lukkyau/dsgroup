````python
# Note that the DataFrame.from_csv has been deprecated - replaced by pd.read_csv
titles=pd.read_csv('pycon-pandas-tutorial/data/titles.csv', index_col=None, encoding='utf-8')
titles.head()

cast=pd.read_csv('pycon-pandas-tutorial/data/cast.csv', index_col=None, encoding='utf-8')
cast.head()
````

````python
# Ask for fewer rows to just remind yourself what the data was
# This creates a new dataframe - so if you've sorted it, you can use it to get the subset you want 
# i.e. it's a proper data slicing/reduction step
h=titles.head(5)
t=titles.tail(5)
h
````

````python
# Each column is called a SERIES (when they're in the dataframe they're called columns)
# Can use square brackets to look up series, or a simple attribute (lower case, no special characters)
h['year']
h.year
````

````python
# You can also do operations on series (e.g. mod10)
h['year'] // 10 * 10
h.year >1985
````

````python
# You can use boolean series above (array of true/falses) to filter out from another data series
h[h.year>1985]

# For multiple conditions, need to use either & (not AND) or | (not OR)
# Note that you need to put both conditions in brackets, otherwise python will try to do 
# "1980 & h.year" FIRST then the booleans around it (due to operator precedence - but not what we want)
h[(h.year<1980) & (h.year >=1990)]
h[(h.year<1980) | (h.year >=1990)]
````

````python
t=titles
t[t.title=="Macbeth"].sort_values('year')
# This creates a dataframe, then we use the SORT_VALUES 
# Can also add on TAIL() and HEAD() to this data frame as well
````

# Exercise 2

- With series, can't use all the standard python STRING methods because of collisions (e.g. can't just do ```series.startswith('hello')```) But can just attach on ```.str``` like - ```series.str.startswith('hello')```. This contains all the string methods.


````python
t=titles
t=t[t.title.str.startswith('Hamlet')]
t
````

- Can use ```titles.year.value_counts()``` that gives you a frequency distribution of the series. Can also use the ```.plot()``` method to create a graph.

````python
t=titles
t=titles.year.value_counts() 
t
titles.year.value_counts().plot() 
# But this is sorted by frequency, so not reasonable
# We know how to sort data (sort_values), but we want to sort by the INDEX
titles.year.value_counts().sort_index()
titles.year.value_counts().sort_index().plot()
````

````python
c=cast
c=c[c.character=="Kermit the Frog"]
c.head()
# .plot automatically uses the index, but you can change it to year for x-axis
c.plot(x='year',y='n',kind='scatter')
````
